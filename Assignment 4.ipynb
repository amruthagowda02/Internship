# Q1

```python
driver=webdriver.Chrome()
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
rank=[]
name=[]
artist=[]
date=[]
view=[]
```

```python
ranks = driver.find_elements(By.XPATH,'//div[@class="mw-parser-output"]/table[2]/tbody/tr/td[1]')
for i in ranks:
    rk = i.text
    rank.append(rk)
names = driver.find_elements(By.XPATH,'//div[@class="mw-parser-output"]/table[2]/tbody/tr/td[2]')
for i in names:
    nm = i.text
    name.append(nm)
artists = driver.find_elements(By.XPATH,'//div[@class="mw-parser-output"]/table[2]/tbody/tr/td[3]')
for i in artists:
    art = i.text
    artist.append(art)
dates = driver.find_elements(By.XPATH,'//div[@class="mw-parser-output"]/table[2]/tbody/tr/td[4]')
for i in dates:
    dt = i.text
    date.append(dt)
views = driver.find_elements(By.XPATH,'//div[@class="mw-parser-output"]/table[2]/tbody/tr/td[5]')
for i in views:
    vw = i.text
    view.append(vw)
```

```python
df=pd.DataFrame({'Rank':rank,'NAME':name,'ARTIST':artist,'VIEWS':date,'DATE':view})
df
```
